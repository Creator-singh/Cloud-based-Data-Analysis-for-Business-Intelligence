# Data Analysis

## Sales Analysis

```python
aggregated_category = amazon.groupby(['Month'])['Amount'].sum().reset_index()
month_order = ['March','April','May','June']
aggregated_category['Month'] = pd.Categorical(aggregated_category['Month'], categories=month_order, ordered=True)
aggregated_category.sort_values('Month').plot(kind='barh',x='Month',y='Amount', xlabel='Total Sales', title='Monthly Sales')
```

As evident from the data, there has been a decline over the past three months. We require the implementation of a marketing campaign to bolster sales. Our approach will involve analyzing both weekend and weekday data to determine the most suitable promotional strategies to pursue.

```python
Weekend = amazon['Day'].isin(['Saturday', 'Sunday'])
Weekday = amazon['Day'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])

df_amazon_weekend = amazon[Weekend].groupby('Month')['Amount'].mean()
df_amazon_weekday = amazon[Weekday].groupby('Month')['Amount'].mean()

concat = pd.concat((df_amazon_weekend,df_amazon_weekday),axis=1,keys=['SalesAverage_Weekend','SalesAverage_Weekday']).reset_index()
concat['Month'] = pd.Categorical(concat['Month'],categories=month_order,ordered=True)
concat.sort_values(by='Month',ascending=True).plot(kind='barh',x='Month',y=['SalesAverage_Weekend','SalesAverage_Weekday'],figsize=(11,5))
```

The average sales on weekdays and weekends in each month do not exhibit significant differences. This indicates that the average customers do not possess any distinct temporal preferences when it comes to shopping.

```python
sns.set_style('whitegrid')
# Resample to monthly sales and plot
monthly_sales = amazon.resample('M', on='Date')['Amount'].sum() 
plt.figure(figsize=(10, 5))
monthly_sales.plot(title='Monthly Sales Over Time') 
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

The sales trend chart displayed indicates a sharp increase in sales from March to April, followed by a peak in April. After reaching this peak, there's a noticeable decline through May and into June. This pattern could suggest a few different scenarios:
- The subsequent decrease could indicate the end of a sales promotion, a seasonal downtrend, or the saturation of a particular market segment following   a period of high demand.
- The peak in April might be due to seasonal factors, specific promotions, or external events that drove a higher volume of sales.nd.

```python
monthly_sales_b2b = amazon[amazon['B2B']].resample('M', on='Date')['Amount'].sum()
monthly_sales_non_b2b = amazon[~amazon['B2B']].resample('M', on='Date')['Amount'].sum()
# Plotting the sales trends over time for B2B transactions
plt.figure(figsize=(10, 5))
monthly_sales_b2b.plot(label='B2B Sales', color='blue') 
monthly_sales_non_b2b.plot(label='Non-B2B Sales', color='orange') 
plt.title('Monthly Sales Trends for B2B and Non-B2B Transactions') 
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

In contrast, the stability of Non-B2B sales implies a steady consumer market without the same level of fluctuation seen in the B2B segment.

```python
# B2B vs. Non-B2B Order Value Analysis
b2b_order_value = amazon.groupby('B2B')['Amount'].mean()
plt.figure(figsize=(10, 5))
b2b_order_value.plot(kind='barh', title='Average Order Value: B2B vs Non-B2B')
plt.ylabel('Is B2B?')
plt.xlabel('Average Order Value')
plt.yticks([0, 1], ['Non-B2B', 'B2B'])
plt.show()
```

The horizontal bar chart visualizing the Average Order Value for B2B versus Non-B2B transactions shows a clear distinction between the two categories.
B2B transactions have a notably higher average order value compared to Non-B2B transactions. This could imply that business customers tend to place larger orders, which might be due to bulk buying or a preference for higher-value transactions when businesses engage with one another. It's also possible that B2B transactions involve a different purchasing dynamic, such as negotiated contracts or recurring orders, which could naturally lead to higher values. The insight gained here could be leveraged to tailor marketing strategies, optimize inventory for B2B needs, or even streamline the sales process to cater more effectively to business clients.

## State wise category analysis.

```python
#Define the color palette for the product categories
colors = {'Top': '#d9d9d9', 'Set': '#969696', 'kurta': '#bdbdbd', 'Western Dress':'orange'}

# Group the data by state and calculate the total sales
sales_by_state = amazon.groupby('ship-state')['Amount'].sum()

# Get the top 5 and bottom 5 states by sales
n_states = len(sales_by_state)
quartiles = pd.qcut(sales_by_state, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
top_states = []
for q in ['Q1', 'Q2', 'Q3', 'Q4']:
    top_states += sales_by_state[quartiles == q].nlargest(5).index.tolist()
    
# Filter the dataframe to include only the top states
top_sales = amazon[amazon['ship-state'].isin(top_states)]

# Group the data by state and product, and calculate the total sales
sales_by_state_product = top_sales.groupby(['ship-state', 'Category'])['Amount'].sum().reset_index()
```

```python
# Get the top 3 products for each quartile
top_products = []
for q in ['Q1', 'Q2', 'Q3', 'Q4']:
    top_products += sales_by_state_product[sales_by_state_product['ship-state'].isin(sales_by_state[quartiles == q].index)].groupby('ship-state').apply(lambda x: x.nlargest(3, 'Amount'))['Category'].tolist()
```

```python
# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(16, 12), sharey=False)

# Create the subplot for quartile 1
q1_sales = sales_by_state_product[sales_by_state_product['ship-state'].isin(sales_by_state[quartiles == 'Q1'].index)]
q1_sales = q1_sales[q1_sales['Category'].isin(top_products)]
sns.barplot(x='ship-state', y='Amount', hue='Category', data=q1_sales, ax=axs[0, 0], palette=colors)
axs[0, 0].legend().set_visible(False)
axs[0, 0].set_title('Sales by Q1 States')
axs[0, 0].set_xlabel('State')
axs[0, 0].set_ylabel('Total Sales')

# Create the subplot for quartile 2
q2_sales = sales_by_state_product[sales_by_state_product['ship-state'].isin(sales_by_state[quartiles == 'Q2'].index)]
q2_sales = q2_sales[q2_sales['Category'].isin(top_products)]
sns.barplot(x='ship-state', y='Amount', hue='Category', data=q2_sales, palette=colors, ax=axs[0, 1])
axs[0, 1].legend().set_visible(False)
axs[0, 1].set_title('Sales by Q2 States')
axs[0, 1].set_xlabel('State')
axs[0, 1].set_ylabel('Total Sales')

# Create the subplot for quartile 3
q3_sales = sales_by_state_product[sales_by_state_product['ship-state'].isin(sales_by_state[quartiles == 'Q3'].index)]
q3_sales = q3_sales[q3_sales['Category'].isin(top_products)]
sns.barplot(x='ship-state', y='Amount', hue='Category', data=q3_sales, palette=colors, ax=axs[1, 0])
axs[1, 0].legend().set_visible(False)
axs[1, 0].set_title('Sales by Q3 States')
axs[1, 0].set_xlabel('State')
axs[1, 0].set_ylabel('Total Sales')

# Create the subplot for quartile 4
q4_sales = sales_by_state_product[sales_by_state_product['ship-state'].isin(sales_by_state[quartiles == 'Q4'].index)]
q4_sales = q4_sales[q4_sales['Category'].isin(top_products)]
sns.barplot(x='ship-state', y='Amount', hue='Category', data=q4_sales, palette=colors, ax=axs[1, 1])
axs[1, 1].legend().set_visible(False)
axs[1, 1].set_title('Sales by Q4 States')
axs[1, 1].set_xlabel('State')
axs[1, 1].set_ylabel('Total Sales')

# Adjust the spacing between subplots
plt.subplots_adjust(hspace=.9, wspace=.21, bottom=.2)

# Rotate the x-axis tick labels
for ax in axs.flat:
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# Add a single legend to the figure
handles, labels = axs[1, 1].get_legend_handles_labels()
fig.legend(handles, labels, loc='center', ncol=len(labels), title='Product Category', fontsize=15, title_fontsize = 15)
    
    
plt.suptitle("Identifying Top 4 Products for Top 5 States in each Revenue Quartile", fontsize = 25)
# Show the figure
plt.show()
```

## Size sales analysis

## Shipping Analysis

```python
shipped_count = amazon["Courier Status"].value_counts()["Shipped"]
cancelled_count = amazon["Courier Status"].value_counts()["Cancelled"]
unknown_count = amazon["Courier Status"].value_counts()["unknown"]
unshipped_count = amazon["Courier Status"].value_counts()["Unshipped"]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
        
status_count = [shipped_count, cancelled_count, unknown_count, unshipped_count]
status_value = ["Shipped", "Cancelled", "Unknown", "Unshipped"]

plt.title("Courier Status of orders")
plt.xlabel("Order Status")
plt.ylabel("No. of orders")
plt.bar(status_value, status_count, color=["darkgreen", "maroon", "yellow", "black"])
addlabels(status_value,status_count)
plt.show()
```

From the above graph we can clearly see that 6872 no. of orders are not known whether they are shipped or not. These huge number can be brought down with proper order tracking and use of blockchain in shipment to convert these orders into profit.

```python
amazon_fulfilment = amazon[amazon['Fulfilment'] == 'Amazon']
amazon_fulfilment = amazon_fulfilment['Status'].value_counts().reset_index()
amazon_fulfilment = amazon_fulfilment.rename(columns={'count':'Amount'})
amazon_fulfilment['composition'] = amazon_fulfilment['Amount']/sum(amazon_fulfilment['Amount'])
amazon_fulfilment = amazon_fulfilment.sort_values(by='composition',ascending=True)

merchant_fulfilment = amazon[amazon['Fulfilment']=='Merchant']
merchant_fulfilment = merchant_fulfilment['Status'].value_counts().reset_index()
merchant_fulfilment = merchant_fulfilment.rename(columns={'count':'Amount'})
merchant_fulfilment['composition'] = merchant_fulfilment['Amount']/sum(merchant_fulfilment['Amount'])
merchant_fulfilment = merchant_fulfilment.sort_values(by='composition',ascending=True)

def add_data_labels_and_adjust_limit(ax, bars):
    for bar in bars:
        xval = bar.get_width()
        yval = bar.get_y() + bar.get_height() / 2
        ax.text(xval, yval, f'{xval:.3f}%', va='center', ha='left', color='black', fontsize=10)
    max_xval = max([bar.get_width() for bar in bars])
    ax.set_xlim(0, max_xval * 1.2)

fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(16,7))
bars1 = axs[0].barh(amazon_fulfilment['Status'],amazon_fulfilment['composition']*100)
axs[0].set(xlabel='Composition',ylabel='Status')
axs[0].set_title('Fulfilment by Amazon')
add_data_labels_and_adjust_limit(axs[0],bars1)

bars2 = axs[1].barh(merchant_fulfilment['Status'],merchant_fulfilment['composition']*100)
axs[1].set(xlabel='Composition',ylabel='Status')
axs[1].set_title('Fulfilment by Merchant')
add_data_labels_and_adjust_limit(axs[1],bars2)

plt.subplots_adjust(wspace=0.6)
plt.show()
```

Fulfilment by Amazon (FBA) boasts a higher rate of successful deliveries compared to Fulfilment by Merchant (FBM). FBM presents a range of potential issues that can arise during the shipping process, making customers more inclined to opt for stores utilizing FBA over FBM.

The risk level associated with FBA is notably lower than that of FBM. This makes FBA an appealing choice for sellers, ensuring that orders reach customers successfully without cancellations or delays.

## Promotion usage Analysis

```python
def update_promotion(promotion_text):
    if 'IN Core Free Shipping' in promotion_text:
        return 'IN Core Free Shipping'
    elif 'Amazon PLCC Free-Financing Universal Merchant' in promotion_text:
        return 'Amazon PLCC Free-Financing Universal Merchant'
    elif 'VPC' in promotion_text:
        return 'VPC Coupon'
    else:
        return promotion_text

amazon['promotion-ids'] = amazon['promotion-ids'].apply(update_promotion)

amazon_promotion = amazon[amazon['Fulfilment']=='Amazon']
amazon_promotion = amazon_promotion['promotion-ids'].value_counts()


merchant_promotion = amazon[amazon['Fulfilment']=='Merchant']
merchant_promotion = merchant_promotion['promotion-ids'].value_counts()

promotion_concat = pd.concat((amazon_promotion,merchant_promotion),axis=1,keys=['qty_amazon_promotion','qty_merchant_promotion']).fillna(0)

promotion_concat.sort_values(by='qty_amazon_promotion',ascending=True).plot(kind='barh')
```

```python
# Adjusting the logic to account for the explicit 'No Promotion' label
amazon['promotion_applied'] = amazon['promotion-ids'] != 'no promotion'
# Now we can re-calculate the average order value with and without promotions applied
promotion_order_value = amazon.groupby('promotion_applied') ['Amount'].mean()

plt.figure(figsize=(10, 5))
promotion_order_value.plot(kind='barh', title='Average Order Value With and Without Promotion')
plt.ylabel('Promotion Applied')
plt.xlabel('Average Order Value')
plt.yticks([0, 1], ['No', 'Yes'])
plt.show()
```
